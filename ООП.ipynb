```python
from datetime import date, datetime
import logging
from abc import ABC, abstractmethod
from typing import Dict, Optional, Type, List
from functools import wraps


# ========== Исключения ==========
class InvalidResourceError(Exception):
    """Ошибка некорректных данных ресурса"""
    pass


class PermissionDeniedError(Exception):
    """Ошибка недостатка прав"""
    pass


class ResourceNotFoundError(Exception):
    """Ошибка отсутствия ресурса"""
    pass


# ========== Настройка логирования ==========
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('library.log'),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)


# ========== Цепочка обязанностей ==========
class SubscriptionRequest:
    """Запрос на изменение подписки"""

    def __init__(self, request_type: str, description: str):
        self.type = request_type  # 'minor', 'price_change', 'admin'
        self.description = description
        self.approved = False
        self.approved_by = None


class SubscriptionHandler(ABC):
    """Базовый обработчик для цепочки"""

    def __init__(self, next_handler: Optional['SubscriptionHandler'] = None):
        self.next_handler = next_handler

    def handle(self, request: SubscriptionRequest) -> None:
        if self.can_handle(request):
            request.approved = True
            request.approved_by = self.__class__.__name__
            logger.info(f"Запрос одобрен {self.__class__.__name__}")
        elif self.next_handler:
            self.next_handler.handle(request)
        else:
            logger.warning("Запрос не может быть обработан")

    @abstractmethod
    def can_handle(self, request: SubscriptionRequest) -> bool:
        pass


class SupportManager(SubscriptionHandler):
    """Менеджер поддержки (обрабатывает простые запросы)"""

    def can_handle(self, request: SubscriptionRequest) -> bool:
        return request.type == 'minor'


class FinanceDepartment(SubscriptionHandler):
    """Финансовый отдел (обрабатывает изменения цен)"""

    def can_handle(self, request: SubscriptionRequest) -> bool:
        return request.type == 'price_change'


class Admin(SubscriptionHandler):
    """Администратор (обрабатывает все запросы)"""

    def can_handle(self, request: SubscriptionRequest) -> bool:
        return True


# ========== Шаблонный метод ==========
class DownloadProcess(ABC):
    """Базовый класс процесса загрузки"""

    def download(self, resource: 'DigitalResource', user: 'User') -> None:
        """Шаблонный метод"""
        self._check_availability(resource)
        self._verify_permissions(user)
        self._download_resource(resource)
        self._confirm_download(resource, user)
        logger.info(f"Ресурс {resource.title} загружен")

    def _check_availability(self, resource: 'DigitalResource') -> None:
        if not resource.is_available:
            raise ResourceNotFoundError("Ресурс недоступен")

    def _verify_permissions(self, user: 'User') -> None:
        if not user.subscription or not user.subscription.is_active():
            raise PermissionDeniedError("Нет активной подписки")

    @abstractmethod
    def _download_resource(self, resource: 'DigitalResource') -> None:
        pass

    def _confirm_download(self, resource: 'DigitalResource', user: 'User') -> None:
        logger.info(f"Подтверждение загрузки для {user.name}")


class EbookDownloadProcess(DownloadProcess):
    """Процесс загрузки электронной книги"""

    def _download_resource(self, resource: 'DigitalResource') -> None:
        print(f"Загружаем книгу {resource.title} в формате EPUB")


class AudiobookDownloadProcess(DownloadProcess):
    """Процесс загрузки аудиокниги"""

    def _download_resource(self, resource: 'DigitalResource') -> None:
        print(f"Загружаем аудиокнигу {resource.title} в формате MP3")


# ========== Декоратор прав доступа ==========
def check_permissions(required_level: str):
    """Декоратор для проверки прав доступа"""

    def decorator(func):
        @wraps(func)
        def wrapper(user: 'User', *args, **kwargs):
            if not user.subscription or not user.subscription.is_active():
                raise PermissionDeniedError("Недостаточно прав")
            if required_level == 'premium' and user.subscription.sub_type != 'premium':
                raise PermissionDeniedError("Требуется премиум подписка")
            return func(user, *args, **kwargs)

        return wrapper

    return decorator


# ========== Фабричные методы ==========
class ResourceFactory:
    @staticmethod
    def create_resource(resource_type: str, resource_id: int, title: str, author: str, file_size: float, **kwargs):
        resource_type = resource_type.lower()
        if resource_type == "ebook":
            return Ebook(
                resource_id=resource_id,
                title=title,
                author=author,
                file_size=file_size,
                page_count=kwargs.get("page_count", 0)  # По умолчанию 0 страниц
            )
        elif resource_type == "audiobook":
            return Audiobook(
                resource_id=resource_id,
                title=title,
                author=author,
                file_size=file_size,
                duration=kwargs.get("duration", 0.0)  # По умолчанию 0 минут
            )
        elif resource_type == "movie":
            return Movie(
                resource_id=resource_id,
                title=title,
                director=author,
                file_size=file_size,
                resolution=kwargs.get("resolution", "1080p")  # По умолчанию разрешение 1080p
            )
        elif resource_type == "music":
            return MusicTrack(
                resource_id=resource_id,
                title=title,
                artist=author,
                file_size=file_size,
                genre=kwargs.get("genre", "Unknown")  # По умолчанию жанр "Unknown"
            )
        else:
            print(f"Неизвестный тип ресурса: {resource_type}")
            return None


# ========== МЕТАКЛАСС ==========
class ResourceMeta(type):
    """Метакласс для автоматической регистрации ресурсов"""
    _registry: Dict[str, Type['DigitalResource']] = {}

    def __new__(cls, name, bases, namespace):
        # Создаем новый класс
        new_class = super().__new__(cls, name, bases, namespace)
        # Регистрируем только конкретные классы (не абстрактные)
        if not getattr(new_class, '__abstractmethods__', None):
            cls._registry[name.lower()] = new_class
        return new_class


# ========== Интерфейсы ==========
class Downloadable(ABC):
    @abstractmethod
    def download(self, user: 'User') -> None:
        """Метод для загрузки ресурса"""
        pass


class Reportable(ABC):
    @abstractmethod
    def generate_report(self) -> str:
        """Метод для генерации отчета"""
        pass


class DigitalResource(Downloadable, Reportable, ABC):
    def __init__(self, resource_id: int, title: str, author: str, file_size: float, is_available: bool = True):
        self._resource_id = resource_id
        self._title = title
        self._author = author
        self._file_size = file_size
        self._is_available = is_available
        self._downloads: List[Dict] = []  # История загрузок

    # ===== РЕАЛИЗАЦИЯ МЕТОДА ИЗ Downloadable =====
    def download(self, user: 'User') -> None:
        """Конкретная реализация метода интерфейса"""
        self._downloads.append({
            'user': user.name,
            'date': datetime.now()
        })
        print(f"Ресурс '{self._title}' загружен пользователем {user.name}")

    # ===== РЕАЛИЗАЦИЯ МЕТОДА ИЗ Reportable =====
    def generate_report(self) -> str:
        """Конкретная реализация метода интерфейса"""
        report = (
            f"Отчёт по ресурсу '{self._title}'\n"
            f"Автор: {self._author}\n"
            f"Всего загрузок: {len(self._downloads)}\n"
        )
        if self._downloads:
            report += "Последние загрузки:\n"
            for download in self._downloads[-3:]:
                report += f"- {download['date']} пользователем {download['user']}\n"
        return report

    @property
    def resource_id(self) -> int:
        return self._resource_id

    @property
    def title(self) -> str:
        return self._title

    @property
    def author(self) -> str:
        return self._author

    @property
    def file_size(self) -> float:
        return self._file_size

    @property
    def is_available(self) -> bool:
        return self._is_available

    @resource_id.setter
    def resource_id(self, value: int):
        self._resource_id = value

    @title.setter
    def title(self, value: str):
        self._title = value

    @file_size.setter
    def file_size(self, value: float):
        self._file_size = value

    @author.setter
    def author(self, value: str):
        self._author = value

    @is_available.setter
    def is_available(self, value: bool):
        self._is_available = value

    @abstractmethod
    def get_resource_info(self) -> str:
        pass

    def __str__(self) -> str:
        return f"Ресурс: {self._title}, Автор: {self._author}"

    def __lt__(self, other) -> bool:
        if isinstance(other, DigitalResource):
            return (self._file_size, self._title) < (other._file_size, other.__title)
        return NotImplemented

    def __gt__(self, other) -> bool:
        if isinstance(other, DigitalResource):
            return (self._file_size, self._title) > (other._file_size, other.__title)
        return NotImplemented


# ========== Подклассы ресурсов ==========
class Ebook(DigitalResource):
    def __init__(self, resource_id: int, title: str, author: str, file_size: float, page_count: int,
                 is_available: bool = True):
        super().__init__(resource_id, title, author, file_size, is_available)
        self.__page_count = page_count

    @property
    def page_count(self) -> int:
        return self.__page_count

    @page_count.setter
    def page_count(self, value: int):
        self.__page_count = value

    def get_resource_info(self) -> str:
        return f"Электронная книга: {self._title}, Страниц: {self.__page_count}"

    def __str__(self) -> str:
        return f"Электронная книга: {self._title}, Автор: {self._author}, Страниц: {self.__page_count}"


class Audiobook(DigitalResource):
    def __init__(self, resource_id: int, title: str, author: str, file_size: float, duration: float,
                 is_available: bool = True):
        super().__init__(resource_id, title, author, file_size, is_available)
        self.__duration = duration

    @property
    def duration(self) -> float:
        return self.__duration

    @duration.setter
    def duration(self, value: float):
        self.__duration = value

    def get_resource_info(self) -> str:
        return f"Аудиокнига: {self._title}, Продолжительность: {self.__duration} минут"

    def __str__(self) -> str:
        return f"Аудиокнига: {self._title}, Автор: {self._author}, Длительность: {self.__duration} мин"


class Movie(DigitalResource):
    def __init__(self, resource_id: int, title: str, director: str, file_size: float, resolution: str,
                 is_available: bool = True):
        super().__init__(resource_id, title, director, file_size, is_available)
        self.__resolution = resolution

    @property
    def resolution(self) -> str:
        return self.__resolution

    @resolution.setter
    def resolution(self, value: str):
        self.__resolution = value

    def get_resource_info(self) -> str:
        return f"Фильм: {self._title}, Разрешение: {self.__resolution}"

    def __str__(self) -> str:
        return f"Фильм: {self._title}, Режиссер: {self._author}, Разрешение: {self.__resolution}"

class MusicTrack(DigitalResource):
    def __init__(self, resource_id: int, title: str, artist: str, file_size: float, genre: str,
                 is_available: bool = True):
        super().__init__(resource_id, title, artist, file_size, is_available)
        self.__genre = genre

    @property
    def genre(self) -> str:
        return self.__genre

    @genre.setter
    def genre(self, value: str):
        self.__genre = value

    def get_resource_info(self) -> str:
        return f"Музыкальный трек: {self._title}, Жанр: {self.__genre}"

    def __str__(self) -> str:
        return f"Музыка: {self._title}, Исполнитель: {self._author}, Жанр: {self.__genre}"

class Subscription:
    def __init__(self, sub_type: str, expiry_date: date):
        self._sub_type = sub_type
        self._expiry_date = expiry_date

    @property
    def sub_type(self) -> str:
        return self._sub_type

    @property
    def expiry_date(self) -> date:
        return self._expiry_date

    def is_active(self) -> bool:
        """Проверяет, активна ли подписка"""
        return self._expiry_date >= datetime.today()

class User:
    def __init__(self, user_id: int, name: str, email: str):
        self._user_id = user_id  # Идентификатор пользователя
        self._name = name  # Имя пользователя
        self._email = email  # Email пользователя
        self._subscription: Optional[Subscription] = None
        self._downloaded_resources: List[DigitalResource] = []

    @property
    def user_id(self) -> int:
        return self._user_id

    @property
    def name(self) -> str:
        return self._name

    @property
    def email(self) -> str:
        return self._email

    @property
    def subscription(self) -> Optional[Subscription]:
        return self._subscription

    def set_subscription(self, sub_type: str, expiry_date: str):
        self._subscription = Subscription(sub_type, expiry_date)

    def download_resources(self, value: DigitalResource):
        if value not in self._downloaded_resources:
            self._downloaded_resources.append(value)

    def remove_resource(self, value: DigitalResource):
        if value in self._downloaded_resources:
            self._downloaded_resources.remove(value)

    def get_downloaded_resources(self) -> List[DigitalResource]:
        return self._downloaded_resources.copy()

# Создаем пользователей
user1 = User(user_id=1, name="Иван Иванов", email="ivan@example.com")
user2 = User(user_id=2, name="Мария Петрова", email="maria@example.com")

# Назначаем подписки пользователям
user1.set_subscription(sub_type="basic", expiry_date=datetime(2024, 12, 31))
user2.set_subscription(sub_type="premium", expiry_date=datetime(2023, 10, 15))

# Создаем ресурсы
ebook = Ebook(resource_id=1, title="Война и Мир", author="Лев Толстой", file_size=2.5, page_count=1225)
audiobook = Audiobook(resource_id=2, title="Мастер и Маргарита", author="Михаил Булгаков", file_size=300.0, duration=600.0)

# Тестируем загрузку ресурсов
try:
    ebook_process = EbookDownloadProcess()
    audiobook_process = AudiobookDownloadProcess()

    print("\nПопытка загрузки электронной книги пользователем 1:")
    ebook_process.download(ebook, user1)  # Успешная загрузка

    print("\nПопытка загрузки аудиокниги пользователем 2:")
    audiobook_process.download(audiobook, user2)  # Успешная загрузка

    print("\nПопытка загрузки электронной книги пользователем 2 (просроченная подписка):")
    ebook_process.download(ebook, user2)  # Ошибка: подписка просрочена

except PermissionDeniedError as e:
    print(f"Ошибка: {e}")

except ResourceNotFoundError as e:
    print(f"Ошибка: {e}")

# Тестируем генерацию отчетов
print("\nГенерация отчетов по ресурсам:")
print(ebook.generate_report())
print(audiobook.generate_report())

# Тестируем фабричный метод для создания ресурсов
print("\nТестирование фабричного метода:")
resource_factory = ResourceFactory()
new_ebook = resource_factory.create_resource(
    "ebook",
    resource_id=3,
    title="Преступление и наказание",
    author="Федор Достоевский",
    file_size=1.8,
    page_count=525
)
if new_ebook:
    print(new_ebook.get_resource_info())
```
