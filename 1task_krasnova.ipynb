5) Поле шахматной доски определяется парой натуральных чисел, каждое из которых не превосходит восьми: первое число — номер вертикали (при счете слева направо), второе — номер горизонтали (при счете снизу вверх). Даны натуральные числа a, b, c, d, e, f, каждое из которых не превосходит восьми. На поле (a, b) расположена белая фигура, на поле (c, d) — черная. Определить, может ли белая фигура пойти на поле (e, f), не попав при этом под удар черной фигуры.
Рассмотреть следующие варианты сочетаний белой и черной фигур:
а) ладья и слон;
б) ферзь и ферзь;
в) ферзь и ладья;


```python
def fun_a(a, b, c, d, e, f):
    if (a == e or b == f): #проверка ладьи
        if (abs(c - e) == abs(d - f)): #проверка слона
            print("не может")
        else:
            print("может")
    else:
        print("ладья так не ходит")

def fun_b(a, b, c, d, e, f):
    if (abs(a - e) == abs(b - f) or (a == e or b == f)):
        if (abs(c - e) == abs(d - f)) or (c == e or d == f):
            print("не может")
        else:
            print("может")
    else:
        print("ферзь так не ходит")

def fun_с(a, b, c, d, e, f):
    if (abs(a - e) == abs(b - f)) or (a == e or b == f):
        if c == e or d == f:
            print("не может")
        else:
            print("может")
    else:
        print("ферзь так не ходит")

a, b = int(input()), int(input())
c, d = int(input()), int(input())
e, f = int(input()), int(input())

variant = input("Enter variant: ")
if (variant == 'a'):
    fun_a(a, b, c, d, e, f)
elif (variant == 'b'):
    fun_b(a, b, c, d, e, f)
else:
    fun_с(a, b, c, d, e, f)
```

```python
from math import gcd

def add_fractions(n1, d1, n2, d2):
    num = n1 * d2 + n2 * d1
    den = d1 * d2
    divisor = gcd(num, den)
    return num // divisor, den // divisor

k = int(input())
numerator, denominator = 1, 1
for i in range(1, k):
    # вместо 1/(1+a_n) дробь denominator/(numerator+denominator)
    numerator, denominator = add_fractions(numerator, denominator, denominator, numerator+denominator)
print(f'{numerator}/{denominator}')
```

```python
from math import sqrt

def distance(point1, point2):
    return sqrt((point1[0]- point2[0])**2+(point1[1]-point2[1])**2)

n = int(input("n: "))
points = []
for i in range(n):
    x, y = int(input("x: ")), int(input("y: "))
    points.append((x, y))

distances = []
for i in range(n):
    for j in range(n):
        if (i!=j):
            distances.append(distance(points[i], points[j]))
print(distances)
```
