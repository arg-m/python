```python

def count_words(s):
    words = s.split()
    frequency = {}
    for word in words:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    return frequency

print(count_words("apple banana apple orange banana apple"))
```

```python
def invert_dict(d):
    result = {}
    for key, value in d.items():
        if value not in result:
            result[value] = [key]
        else:
            result[value].append(key)
    return result

print(invert_dict({'a': 1, 'b': 2, 'c': 1}))
```

```python
def merge_dicts(d1, d2):
    result = d1.copy()
    for key, value in d2.items():
        if key in result:
            result[key] += value
        else:
            result[key] = value
    return result

print(merge_dicts({'a': 1, 'b': 2}, {'b': 3, 'c': 4}))
```

```python
def filter(d, threshold):
    result = {}
    for key, value in d.items():
        if value > threshold:
            result[key] = value
    return result

print(filter({'a': 1, 'b': 3, 'c': 2}, 2))
```

```python

def max_value(d):
    mx = 0;
    for key, value in d.items():
        mx = max(value, mx);
    return mx

print(max_value({'a': 1, 'b': 3, 'c': 2}))
```

```python
def min_value(d):
    mn = next(iter(d.values()));
    for key, value in d.items():
        mx = min(value, mn);
    return mn

print(min_value({'a': 1, 'b': 3, 'c': 2}))
```

```python

def group_by_key_length(d):
    result = {}
    for k in d.keys():
        length = len(k)
        if length not in result:
            result[length] = []
        result[length].append(k)
    return result

print(group_by_key_length({'a': 4, 'abc': 2, 'abcd': 3}))

```

```python

def combine_values(d1, d2):
    combined = {}
    for key in set(d1) | set(d2):
        combined[key] = []
        if key in d1:
            combined[key].append(d1[key])
        if key in d2:
            combined[key].append(d2[key])
    return combined

print(combine_values({'a': 1, 'b': 2}, {'b': 3, 'c': 4}))
```
